```python
#import libraries
import numpy as np
import pandas as pd
import textwrap
from pprint import pprint


from transformers import pipeline
```

```python
#read in data
df=pd.read_csv('bbc_text_cls.csv')
df.head()
```

```python
df.labels.value_counts()
```

```python
#print out the unique set of labels
labels=set(df.labels)
labels
```

```python
#pick a label
label='business'
```

```python
#select business text
texts=df[df['labels']==label]['text']
texts.head()
```

```python
np.random.seed(1234)
```

```python
#let choose a random documents
i=np.random.choice(texts.shape[0])
doc=texts.iloc[i]
```

```python
#let print it out
print(textwrap.fill(doc, replace_whitespace=False, fix_sentence_endings=True))
```

```python
mlm=pipeline('fill-mask')
```

```python
mlm('Business confidence dips in  <mask>')
```

```python
text='Business confidence among Japanese' + \
'manufacturers has weakened for the first time since <mask> 2003, the' + \
'quarterly Tankan survey has found'

mlm(text)
```

```python
text= ' Japans economy grew by just 0.1% in the three months to September, ' + \
 'according revised data issued this month.  With the <mask> slowing, ' + \
 ' the world s second biggest economy is now expected grow by 0.2% in 2004' 

pprint(mlm(text))
```
