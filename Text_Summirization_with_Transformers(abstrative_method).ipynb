```python
#import libraries
import pandas as pd
import numpy as np
import textwrap
from transformers import pipeline
```

```python
#read in the data
df=pd.read_csv('bbc_text_cls.csv')
```

```python
df.head()
```

```python
#select the business category and pick an article at random
doc=df[df.labels=='business']['text'].sample(random_state=42)
```

```python
#wrap the text to fit the page
def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
#let test our function
print(wrap(doc.iloc[0]))
```

```python
#instantiate our pipeline object, by default distilbart is used which is a seq2seq model
summarizer=pipeline("summarization")
```

```python
#use our model on out text. The title is ignored
summarizer(doc.iloc[0].split("\n", 1)[1])
```

```python
#let create a function that text in a doc and print out the summary
def print_summary(doc):
  result=summarizer(doc.iloc[0].split("\n", 1)[1])
  print(wrap(result[0]['summary_text']))
```

```python
print_summary(doc)
```
